```python
df.isnull().any()
```

```python
import altair as alt


alt.Chart(df).mark_bar().encode(
    alt.X('confirmed_date'),
    alt.Y('n'),
    
    alt.Color('infection_reason'),
    alt.OpacityValue(2),
    tooltip = [alt.Tooltip('sex'),
               alt.Tooltip('state'),
               alt.Tooltip('province'),
               alt.Tooltip('infection_reason'),
               alt.Tooltip('country')
              ]
    
).interactive().properties(
    width=700,
    height=500
)
```

```python
import altair as alt


alt.Chart(df).mark_bar().encode(
    alt.X('province'),
    alt.Y('age'),
    
    alt.Color('sex'),
    alt.OpacityValue(2),
    tooltip = [alt.Tooltip('sex'),
               alt.Tooltip('state'),
               alt.Tooltip('province'),
               alt.Tooltip('infection_reason'),
               alt.Tooltip('country')
              ]
    
).interactive().properties(
    width=800,
    height=500
)
```

```python
import altair as alt


alt.Chart(df).mark_boxplot().encode(
    alt.X('sex'),
    alt.Y('age'),
    
    alt.Color('infection_reason'),
    alt.OpacityValue(2),
    tooltip = [alt.Tooltip('sex'),
               alt.Tooltip('state'),
               alt.Tooltip('province'),
               alt.Tooltip('infection_reason'),
               alt.Tooltip('country')
               
              ]
    
).interactive().properties(
    width=200,
    height=500
)
```
